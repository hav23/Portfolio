```python
import matplotlib.pyplot as plt
import pandas as pd
youtube_views_data = [543,689,258,401,724,689,350]
days = range(1,8)

plt.plot(days, youtube_views_data, label = "Youtube views", color = "blue", marker = "o", linestyle = "dashed", linewidth = 2)
plt.xlabel("Weekdays")
plt.ylabel("Views")
plt.legend(loc="upper left")
plt.title("Youtube views for given youtube channel")
plt.show()
```

```python
youtube_views_data = [543,689,258,401,724,689,350]
facebook_views_data = [123,342,700,304,405,650,325]
twitter_views_data = [202,209,176,415,824,389,550]
days = range(1,8)

plt.plot(days, youtube_views_data, label = "Youtube views", color = "red", marker = "o", linestyle = "dashed", linewidth = 2)
plt.plot(days, facebook_views_data, label = "Facebook views", color = "blue", marker = "o", linestyle = "dashed", linewidth = 2)
plt.plot(days, twitter_views_data, label = "Twitter views", color = "orange", marker = "o", linestyle = "dashed", linewidth = 2)
plt.xlabel("Weekdays")
plt.ylabel("Views")
plt.legend(loc="upper left")
plt.title("Youtube views for given youtube channel")
plt.show()
```
