```python
# Import libraries
import mysql.connector as connection
import pandas as pd
#import plotly.express as px
#import matplotlib.pyplot as plt
#import os
#import sqlalchemy as sa
pd.options.display.max_columns = 100000
```

```python
try:
    # Get user input for MySQL connection details
    host = input("Enter MySQL host (default is localhost): ") or "localhost"
    username = input("Enter MySQL username (default is root): ") or "root"
    password = input("Enter MySQL password (default is root): ") or "root"
    database = input("Enter MySQL database (default is world): ") or "world"
    table = input("Enter table name (default is city): ") or "city"
    column = input("Enter column name (default is *): ") or "*"

    # Establish a connection to the MySQL database

    mydb = connection.connect(host=host, database = database,user=username, passwd=password,use_pure=True)
    query = f"Select {column} from {table};"
    result_dataFrame = pd.read_sql(query,mydb)
    mydb.close() #close the connection
    print('MySQLfetced data successfully')
except Exception as e:
    mydb.close()
    print(str(e))
```
